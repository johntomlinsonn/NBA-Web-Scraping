```python
import requests

web_url = "https://www.basketball-reference.com/leagues/NBA_2024.html"

data = requests.get(web_url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.text)
caption = soup.find("caption", string="Per Game Stats Table")
teams = caption.find_parent("table")

links = teams.find_all("a")

temp = []

for i in links:
    temp.append(i.get("href"))
links = temp

i = 0
while i < len(links):
    if not("/teams/" in links[i]):
        links.remove(i)
    else:
        i+=1
    
```

```python
temp = []
for i in links:
    temp.append(i[7:10])
for i in range(len(links)):
    links[i] = f"https://www.basketball-reference.com/teams/{temp[i]}/2024/gamelog/"
links
```
