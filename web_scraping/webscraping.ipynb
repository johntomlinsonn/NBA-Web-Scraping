```python
import itertools
import numpy as np

def get_averages(list_teams):
    global csv
    teams = list_teams

    for team in range(len(teams)):
        df = csv.iloc[teams[team][0]:teams[team][1]]
        columns = df.columns[11:53]
        
        # Convert DataFrame to NumPy array for faster operations
        data = df[columns].apply(pd.to_numeric, errors='coerce').fillna(0).to_numpy()
        dates = df["Date"].to_numpy()
        
        for i in range(7, len(df)):
            if dates[i] == "Date" or "-" in df["Rslt"].iloc[i]:
                continue
            else:
                stats = np.zeros(42)
                count = 0
                for j in range(i - 7, i):
                    if dates[j] == "Date":
                        continue
                    else:
                        stats += data[j].astype(np.float64)
                        count += 1
                
                if count > 0:
                    stats /= count
                
                avg_columns = ["avg-" + col for col in columns]
                for k, col in enumerate(avg_columns):
                    csv.at[teams[team][0] + i, col] = stats[k]
                
        print(f"Team {team} averages calculated and applied.")
    return csv

def get_teams_list(all_games):
    teams = []
    team = all_games["Team"].iloc[0]
    start = 0
    for i in range(1, len(all_games)):
        if all_games.at[i, "Team"] != team and not "-" in all_games.at[i, "Rslt"]:
            teams.append([start, i])
            start = i
            team = all_games["Team"].iloc[start]
    teams.append([start, len(all_games)])  # Add the last team
    print(len(teams))
    return teams

list_teams = get_teams_list(csv)
all_avgs = get_averages(list_teams)
print("done")
```

```python
def clean_csv(precsv):
    avg_columns = [col for col in precsv.columns if col.startswith("avg-")]
    rows_to_drop = []

    for i in range(len(precsv)):
        if precsv.at[i, "Date"] == "Date" or "-" in precsv.at[i, "Rslt"]:
            rows_to_drop.append(i)

    precsv.drop(rows_to_drop, inplace=True)
    return precsv
all_avgs = clean_csv(csv)
```
